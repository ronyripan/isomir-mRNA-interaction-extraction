## marking ovelapping mRNA-miRNA portion sequences

```python
mrna_mirna_df['gap'] = 0
#mrna_mirna_df['gap'] = mrna_mirna_df['gap'].astype(int)
```

```python
overlap_condition = (((mrna_mirna_df['qstart_mRNA'] >= mrna_mirna_df['qstart_miRNA']) & (mrna_mirna_df['qstart_mRNA'] <= mrna_mirna_df['qend_miRNA'])) | 
                     ((mrna_mirna_df['qstart_miRNA'] >= mrna_mirna_df['qstart_mRNA']) & (mrna_mirna_df['qstart_miRNA'] <= mrna_mirna_df['qend_mRNA'])))

mrna_mirna_df.loc[overlap_condition, 'gap'] = -1
```

```python
gap_condition_1 = ((mrna_mirna_df['qend_miRNA'] <= mrna_mirna_df['qstart_mRNA']) & (mrna_mirna_df['gap'] != -1))

mrna_mirna_df.loc[gap_condition_1, 'gap'] = mrna_mirna_df['qstart_mRNA'] - mrna_mirna_df['qend_miRNA'] - 1

gap_condition_2 = ((mrna_mirna_df['qend_mRNA'] <= mrna_mirna_df['qstart_miRNA']) & (mrna_mirna_df['gap'] != -1))

mrna_mirna_df.loc[gap_condition_2, 'gap'] = mrna_mirna_df['qstart_miRNA'] - mrna_mirna_df['qend_mRNA'] - 1
```

```python
mrna_mirna_df = mrna_mirna_df[mrna_mirna_df['gap'] <= 4] #allowing maximum 4nt gap
mrna_mirna_df.reset_index(drop=True, inplace=True)
#mrna_mirna_df
```

```python
start1 = mrna_mirna_df['qstart_mRNA']
end1 = mrna_mirna_df['qend_mRNA']
start2 = mrna_mirna_df['qstart_miRNA']
end2 = mrna_mirna_df['qend_miRNA']

# Overlap calculations
overlap1 = np.where(
    (start2 >= start1) & (start2 <= end1),
    np.minimum(end1, end2) - start2 + 1,
    0
)

overlap2 = np.where(
    (start1 >= start2) & (start1 <= end2),
    np.minimum(end1, end2) - start1 + 1,
    0
)

# Combine the overlap calculations
mrna_mirna_df['overlap'] = np.maximum(overlap1, overlap2)
```

```python
mrna_mirna_df = mrna_mirna_df[mrna_mirna_df['overlap'] <= 4] #allowing maximum 4 nucleotide overlap
mrna_mirna_df.reset_index(drop=True, inplace=True)
#mrna_mirna_df
```

```python
def read_fasta(fasta_file):
    """
    Reads a FASTA file and returns a dictionary of sequences.

    Args:
    fasta_file (str): Path to the FASTA file.

    Returns:
    dict: A dictionary with sequence identifiers as keys and sequences as values.
    """
    sequences = {}
    with open(fasta_file, 'r') as file:
        sequence_id = ""
        sequence = ""
        for line in file:
            line = line.strip()
            if line.startswith(">"):
                if sequence_id:
                    sequences[sequence_id] = sequence
                    sequence = ""
                sequence_id = line[1:]  # Remove '>' character
                sequence_id = sequence_id.split()[0] #keeping only miRNA name.
            else:
                sequence += line
                sequence = sequence.replace("U", "T") #replacing U with T since blast replaces U with T
                
        if sequence_id:
            sequences[sequence_id] = sequence
    return sequences
```

```python
def fastq_quality_to_scores(quality_string):
    """
    Convert a FASTQ quality string to a list of quality scores.

    Args:
    quality_string (str): A string of ASCII characters representing quality scores.

    Returns:
    List[int]: A list of numerical quality scores.
    """
    return [ord(char) - 33 for char in quality_string]


import gzip

def read_fastq(file_path):
    sequences = {}
    with gzip.open(file_path, 'rt') as file:
        while True:
            try:
                seq_id = file.readline().strip()  # Sequence identifier
                seq = file.readline().strip()     # Sequence
                plus = file.readline().strip()    # Plus line
                qual = file.readline().strip()    # Quality scores
                
                if not seq_id:
                    break
                
                if seq_id.startswith('@') and plus.startswith('+'):
                    # Strip the '@' from the sequence identifier
                    seq_id = seq_id[1:].split()[0]
                    sequences[seq_id] = fastq_quality_to_scores(qual)
                else:
                    raise ValueError("Incorrect FASTQ format")
            
            except Exception as e:
                print(f"Error: {e}")
                break
                
    return sequences
```

```python
def check_quality(qual_seq):
    for i in range(len(qual_seq)):
        if qual_seq[i] < 30:
            return 1
    return 0

def get_qual_flag(seq_id, i):
    qual_seq = quality_sequneces[seq_id]
    qual_flag = 0
    if mrna_mirna_df['qstart_miRNA'][i] >= mrna_mirna_df['qend_mRNA'][i]: #miRNA is on the right of mRNA
        end_ind = mrna_mirna_df['qlen_miRNA'][i]
        if mrna_mirna_df['gap'][i] <= 0:
            qual_flag = check_quality(qual_seq[mrna_mirna_df['qend_miRNA'][i]-1:end_ind])
        else:
            start_ind = mrna_mirna_df['qstart_miRNA'][i] - int(mrna_mirna_df['gap'][i]) - 1
            qual_flag = check_quality(qual_seq[start_ind:mrna_mirna_df['qstart_miRNA'][i]])

            if qual_flag == 0: 
             qual_flag = check_quality(qual_seq[mrna_mirna_df['qend_miRNA'][i] - 1:end_ind])

    else: # miRNA is on the left of mRNA
        start_ind = 0
        if mrna_mirna_df['gap'][i] <= 0:
            qual_flag = check_quality(qual_seq[start_ind:mrna_mirna_df['qstart_miRNA'][i]])
        else:
            end_ind = mrna_mirna_df['qend_miRNA'][i] + int(mrna_mirna_df['gap'][i])
            qual_flag = check_quality(qual_seq[mrna_mirna_df['qend_miRNA'][i] - 1:end_ind])
            if qual_flag == 0:
                qual_flag = check_quality(qual_seq[start_ind:mrna_mirna_df['qstart_miRNA'][i]])

    return qual_flag
```

```python
def mismatch(chim_mirna, ori_mirna, quality_score):
    if len(chim_mirna) != len(quality_score):
        raise ValueError("The length of chim_mirna does not match the length of quality_sequences")
    
    if len(chim_mirna) != len(ori_mirna):
        raise ValueError("The length of chim_mirna does not match the length of ori_mirna")
    
    for i in range(len(chim_mirna)):
        if chim_mirna[i] != ori_mirna[i]:
            if quality_score[i] < 30:
                return 1
    return 0
```

## Categorizing isomiRNAs into different types

```python
import re
import pandas as pd
```

```python
isomir_df = pd.read_csv(r"C:\Users\16892\OneDrive - University of Central Florida\BIOINFORMATICS LAB\isomir_dataset_paper\outputs\eclip\SRR18280895\R1\isomir.csv")
```

```python
def check_mismatch_positon(chim_mirna, ori_mirna):
    if len(chim_mirna) != len(ori_mirna):
        raise ValueError("The length between chim_mirna and ori_mirna is not equal")
    
    for i in range(len(chim_mirna)):
        if chim_mirna[i] != ori_mirna[i]:
            return 1
    return 0

def get_start_end_ind(chim_mirna, ori_mirna):
    match = re.search(ori_mirna, chim_mirna)
    if match:
        start_index = match.start()
        end_index = match.end() - 1
        return start_index, end_index
    else:
        raise ValueError("chim_mirna does not have ori_mirna")
    

isomir_df['isomir_type'] = ""

for i in range(len(isomir_df)):
    chim_mirna = isomir_df['final_seq'][i] # isomiRNA from chimeric sequence
    ori_mirna = sequences[isomir_df['saccver_miRNA'][i]] #original miRNA sequence from miRBASE
    ori_mirna_portion = isomir_df['sseq_miRNA'][i] #aligned miRNA portion from original miRNA sequence
    mismatch = isomir_df['mismatch'][i]
    front_distance = isomir_df['sstart_miRNA'][i] - 1
    back_distance = isomir_df['slen_miRNA'][i] - isomir_df['send_miRNA'][i]
    start_ind = isomir_df['sstart_miRNA'][i]
    end_ind = isomir_df['send_miRNA'][i]
    strand_info = isomir_df['saccver_miRNA'].iloc[i][-2:]
    type = []

    if len(chim_mirna) == len(ori_mirna):
        if mismatch == 1:
            type.append("SNP")
        else:
            type.append("MNP")

    elif len(chim_mirna) > len(ori_mirna):
        chim_start_ind, chim_end_ind = get_start_end_ind(chim_mirna, ori_mirna)
        front_add = chim_start_ind - 0
        back_add = len(chim_mirna) - chim_end_ind - 1

        if front_add != 0:
            if strand_info == '5p':
                type.append("5p-addition")
            else:
                type.append("3p-addition")

        if back_add != 0:
            if strand_info == '5p':
                type.append("3p-addition")
            else:
                type.append("5p-addition")

        if (front_add > 4) & (back_add > 4):
            new_seq = chim_mirna[chim_start_ind-4:chim_end_ind+4+1]
            isomir_df.at[i, 'final_seq'] = new_seq
        else:
            if front_add > 4:
                new_seq = chim_mirna[chim_start_ind-4:]
                isomir_df.at[i, 'final_seq'] = new_seq

            if back_add > 4:
                new_seq = chim_mirna[:chim_end_ind+4+1]
                isomir_df.at[i, 'final_seq'] = new_seq 
    else:
        if start_ind != 1:
            if strand_info == '5p':
                type.append("5p-deletion")
            else:
                type.append("3p-deletion")
        
        if check_mismatch_positon(chim_mirna[0:4-start_ind+1], ori_mirna_portion[0:4-start_ind+1]) == 1:
            if strand_info == '5p':
                type.append('5p-replacement')
            else:
                type.append('3p-replacement')

        if back_distance != 0:
            if strand_info == '5p':
                type.append('3p-deletion')
            else:
                type.append("5p-deletion")

        if check_mismatch_positon(chim_mirna[-(4-back_distance):], ori_mirna_portion[-(4-back_distance):]) == 1:
            if strand_info == '5p':
                type.append('3p-replacement')
            else:
                type.append('5p-replacement')
            
    isomir_df.at[i, 'isomir_type'] = type            
```
